```python
import pandas as pd
```

```python
actorsdf = pd.read_csv("actors.csv")
crewdf = pd.read_csv("crew.csv")
genresdf = pd.read_csv("genres.csv")
languagesdf = pd.read_csv("languages.csv")
moviesdf = pd.read_csv("movies.csv")
studiosdf = pd.read_csv("studios.csv")
themesdf = pd.read_csv("themes.csv")
```

```python
import re

def clean_title(title):
    return re.sub("[^a-zA-Z0-9 ]", "", title)
```

```python
mylist = ["Writer","Director","Producer"]
crewdf = crewdf[crewdf["role"].isin(mylist)]
grouped_crewdf = crewdf.groupby("id", as_index=False).agg({"name": lambda x: " ".join(x.astype(str))})
grouped_crewdf
```

```python
grouped_crewdf = grouped_crewdf.rename(columns={"name":"crew"})
grouped_crewdf
```

```python
grouped_actorsdf = actorsdf.groupby("id", as_index=False).agg(lambda x: " ".join(x.astype(str)))
grouped_actorsdf = grouped_actorsdf.rename(columns={"name":"actors"})
grouped_actorsdf.head(5)
```

```python
grouped_genredf = genresdf.groupby("id", as_index=False).agg(lambda x: " ".join(x.astype(str)))
grouped_genredf.head(5)
```

```python
mylang = ["Language","Primary language"]
languagesdf = languagesdf[languagesdf["type"].isin(mylang)]
languagesdf = languagesdf.drop(columns="type")
languagesdf.head(10)
```

```python
moviesdf.head(10)
```

```python
grouped_studiodf = studiosdf.groupby("id", as_index=False).agg(lambda x: " ".join(x.astype(str)))
grouped_studiodf.head(10)
```

```python
#id is not the index when setting as _index=false and agg adds the x which Represents the grouped values for each column.
# For each group, x will be a Series containing all values in that group for a particular column.

grouped_themesdf = themesdf.groupby('id',as_index=False).agg(lambda x: " ".join(x.astype(str)))
grouped_themesdf.head(10)
```

```python
# dfs grouped_themesdf,grouped_studiodf,moviesdf,languagesdf,grouped_actorsdf,grouped_crewdf

print(grouped_themesdf.shape)
print(grouped_studiodf.shape)
print(moviesdf.shape)
print(languagesdf.shape)
print(grouped_actorsdf.shape)
print(grouped_crewdf.shape)
print(grouped_genredf.shape)
```

```python
# merginging 
merged_df = pd.merge(moviesdf,languagesdf, how="left", on="id")
merged_df = pd.merge(merged_df,grouped_crewdf, how="left", on="id")
merged_df = pd.merge(merged_df, grouped_genredf,how="left",on="id")
merged_df = pd.merge(merged_df, grouped_actorsdf, how="left", on="id")
merged_df = pd.merge(merged_df, grouped_studiodf, how="left", on="id")
merged_df = pd.merge(merged_df, grouped_themesdf, how="left", on="id")
merged_df
```

```python
merged_df.dropna(subset=["id"], inplace=True)
merged_df
```

```python
# "name","tagline","description","rating","theme"
merged_df["useless"] = merged_df[['description']].isna().all(axis=1)
merged_df[merged_df["useless"] == True].head(20)

cleaned_df = merged_df[merged_df["useless"] == False].reset_index()
cleaned_df.tail(10)
```

```python
cleaned_df.shape
```

```python
cleaned_df = cleaned_df.drop(cleaned_df[cleaned_df["date"] > 2025].index).reset_index()
cleaned_df
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cleaned_df["narrative"] = cleaned_df["tagline"].fillna("") + " " + cleaned_df["description"].fillna("") + " " + cleaned_df["genre"].fillna("") + " " + cleaned_df["theme"].fillna("")
cleaned_df.head(1)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2))

tfidf = vectorizer.fit_transform(cleaned_df["narrative"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(descrip):
    query_vec = vectorizer.transform([descrip])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity,-10)[-10:]
    results = cleaned_df.iloc[indices][::1]
    return results
```

```python
descrip = input("enter movie description")
search_results = search(descrip)
df = pd.DataFrame(search_results)
df
```
