```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("cleaned_movie_df.csv")
df.tail(10)
```

```python
rows_to_drop = df[df["name"].str.contains("Untitled", na=False)].index
df2 = df.drop(rows_to_drop).reset_index(drop=True)
df2.to_csv("cleaned_movie__final_df.csv", index=False)
```

```python
df2["combined"] = df2["tagline"].fillna("") + " " + df2["description"].fillna("") + " " + df2["theme"].fillna("") + " " + df2["genre"].fillna("") + " " + df2["studio"].fillna("")
df2.head(2)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(df2["combined"])
```

```python
def search():
    title = input("Enter movie name:")
    list_of_all_titles = df2["name"].tolist()
    find_close_match = difflib.get_close_matches(title,list_of_all_titles)
    close_match = find_close_match[0] 
    return close_match
```

```python
title = search()
title
```

```python
matched_row = df2[df2["name"] == title]["combined"].values[0]
matched_row
#search_query = vectorizer.transform(df[df["name"] == "close_match"]["combined"].values[0])
```

```python
query_vec = vectorizer.transform([matched_row])
similarity = cosine_similarity(query_vec, feature_vectors).flatten()
indices = np.argpartition(similarity, -10)[-10:]
results = df.iloc[indices][::-1]
results
```
